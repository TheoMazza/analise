```python
import statistics
import matplotlib.pyplot as plt

alturas = [182, 180, 180, 180, 178, 160, 183, 167, 176, 184,
           174, 174, 173, 183, 175, 180, 164, 174, 173, 170,
           181, 178, 163, 173, 172, 173, 162, 178, 179, 166,
           182, 182, 190, 166]

media = statistics.mean(alturas)
print("Média:", media)

mediana = statistics.median(alturas)
print("Mediana:", mediana)

maximo = max(alturas)
print("Valor máximo:", maximo)

minimo = min(alturas)
print("Valor mínimo:", minimo)

variancia = statistics.variance(alturas)
print("Variância:", variancia)

desvio_padrao = statistics.stdev(alturas)
print("Desvio Padrão:", desvio_padrao)

plt.hist(alturas, bins=10, edgecolor='black')
plt.title('Histograma de Alturas')
plt.xlabel('Altura')
plt.ylabel('Frequência')
plt.show()

plt.figure()
plt.hist(alturas, bins=10, density=True, alpha=0.5, color='g', edgecolor='black')
plt.title('Histograma e Curva de Densidade de Alturas')
plt.xlabel('Altura')
plt.ylabel('Densidade')

mu = media
sigma = desvio_padrao
x = [x/10 for x in range(1600, 2000)]
y = [1/(sigma * (2 * 3.14) ** 0.5) * 2.7 ** (-(i - mu) ** 2 / (2 * sigma ** 2)) for i in x]
plt.plot(x, y, 'r', linewidth=2)
plt.show()
```
